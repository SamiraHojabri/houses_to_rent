```python
df['floor'] = df['floor'].replace('-',0)
df
```

```python
df['floor'] = df['floor'].astype(int)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['animal'] = le.fit_transform(df['animal'])
df['city'] = le.fit_transform(df['city'])
df['furniture'] = le.fit_transform(df['furniture'])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
target = df['rent amount (R$)']
data = df.drop(columns = ['rent amount (R$)','total (R$)'])
data
```

```python

x_train,x_test,y_train,y_test = train_test_split(data,target,test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
predicted = lr.predict(x_test)

plt.scatter(y_test,predicted)
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_test,predicted))
```

```python
s1 = np.array([[4,100,2,1,2,2,0,0,100,500,10]]) # just an example for a test
lr.predict(s1)
```
