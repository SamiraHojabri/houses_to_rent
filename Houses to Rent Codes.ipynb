```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['animal'] = le.fit_transform(df['animal'])
df['city'] = le.fit_transform(df['city'])
df['furniture'] = le.fit_transform(df['furniture'])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
target = df['rent amount (R$)']
data = df.drop(columns = ['rent amount (R$)','total (R$)'])
data
```

```python

x_train,x_test,y_train,y_test = train_test_split(data,target,test_size = 0.3)
#ba han dastoor darim dadehaye train(70% dadehaye kol)ro + dadehaye target midim be algorythm
#jahate modelsazi
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python

predicted = lr.predict(x_test)#in khat miad dadehaye x_test ro migire va predict mikone baramoon
plt.scatter(y_test,predicted)#in dastoor miad oonayee ke model predict karde(predicted) va dadehaye vaghei
#ke dashtim khodemoon az ghabl(y_train) ro baramoon moghayese mikone ke cheghad model dorost amak mikone
#dar vaghe deghat model ro be ma mide in dastoor
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,predicted))
```
